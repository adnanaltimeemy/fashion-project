```python
# Step 1: Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display settings
pd.set_option("display.max_columns", None)

print("Libraries loaded successfully.")
```

```python
# Step 2: Create luxury brand dataset

data = {
    "Brand": ["Louis Vuitton", "Prada", "Dior"],
    
    # Financial metrics (approximate realistic values)
    "Revenue_Billion_EUR": [20.0, 5.43, 15.0],
    
    "Operating_Margin_Percent": [35, 19, 22],
    
    "Net_Profit_Billion_EUR": [7.0, 0.839, 3.3],
    
    "Profit_Margin_Percent": [
        (7.0/20.0)*100,
        (0.839/5.43)*100,
        (3.3/15.0)*100
    ],
    
    # Demographics
    "Avg_Customer_Age": [38, 32, 36],
    
    "Female_Percent": [65, 70, 68],
    
    "Male_Percent": [35, 30, 32],
    
    "Asia_Percent": [40, 45, 38],
    
    "Europe_Percent": [30, 25, 32],
    
    "NorthAmerica_Percent": [25, 20, 25],
    
    "Other_Percent": [5, 10, 5]
}

df = pd.DataFrame(data)

df
```

```python
# Step 3: Save dataset

df.to_csv("luxury_brands_dataset.csv", index=False)

print("Dataset saved successfully.")
```

```python
# Step 4: Load dataset

df = pd.read_csv("luxury_brands_dataset.csv")

df.head()
```

```python
# Step 5: Plot revenue comparison

plt.figure()

plt.bar(df["Brand"], df["Revenue_Billion_EUR"])

plt.title("Revenue Comparison")
plt.xlabel("Brand")
plt.ylabel("Revenue (Billion EUR)")

plt.show()
```

```python
# Step 6: Profit margin comparison

plt.figure()

plt.bar(df["Brand"], df["Profit_Margin_Percent"])

plt.title("Profit Margin Comparison")
plt.xlabel("Brand")
plt.ylabel("Profit Margin (%)")

plt.show()
```

```python
# Step 7: Operating margin comparison

plt.figure()

plt.bar(df["Brand"], df["Operating_Margin_Percent"])

plt.title("Operating Margin Comparison")
plt.xlabel("Brand")
plt.ylabel("Operating Margin (%)")

plt.show()
```

```python
# Step 8: Customer age comparison

plt.figure()

plt.bar(df["Brand"], df["Avg_Customer_Age"])

plt.title("Average Customer Age")
plt.xlabel("Brand")
plt.ylabel("Age")

plt.show()
```

```python
# Step 9: Gender demographics

df_gender = df[["Brand", "Female_Percent", "Male_Percent"]]

df_gender.set_index("Brand").plot(kind="bar")

plt.title("Gender Demographics")
plt.ylabel("Percent")

plt.show()
```

```python
# Step 10: Regional demographics

df_region = df[[
    "Brand",
    "Asia_Percent",
    "Europe_Percent",
    "NorthAmerica_Percent",
    "Other_Percent"
]]

df_region.set_index("Brand").plot(kind="bar")

plt.title("Regional Demographics")
plt.ylabel("Percent")

plt.show()
```

```python
# Step 11: Statistical summary

df.describe()
```

```python
# Step 12: Find highest profit margin

highest_profit = df.loc[df["Profit_Margin_Percent"].idxmax()]

print("Most profitable brand:")
print(highest_profit)
```

```python
# Step 13: Sort brands by revenue

df_sorted = df.sort_values(by="Revenue_Billion_EUR", ascending=False)

df_sorted
```

```python
# Step 14: Correlation analysis

df.corr(numeric_only=True)
```

```python
# Step 15: Scatter plot

plt.figure()

plt.scatter(df["Revenue_Billion_EUR"], df["Profit_Margin_Percent"])

for i, brand in enumerate(df["Brand"]):
    plt.text(df["Revenue_Billion_EUR"][i],
             df["Profit_Margin_Percent"][i],
             brand)

plt.title("Profit Margin vs Revenue")
plt.xlabel("Revenue (Billion EUR)")
plt.ylabel("Profit Margin (%)")

plt.show()
```
