```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
brand_df = pd.read_csv("fashion_sales.csv")

# Clean column names
brand_df.columns = brand_df.columns.str.strip().str.lower()

# Print column names
print("Column names:", brand_df.columns.tolist())

# Preview dataset
print("\nDataset preview:")
print(brand_df.head())

# Automatically detect likely brand and revenue columns
brand_column = brand_df.columns[0]
revenue_column = brand_df.columns[-1]

print(f"\nUsing '{brand_column}' as Brand column")
print(f"Using '{revenue_column}' as Revenue column")

# Group and sum
sales_comparison = brand_df.groupby(brand_column)[revenue_column].sum()

# Print results
print("\nTotal Revenue by Brand:")
print(sales_comparison)

# Plot
plt.figure(figsize=(8,5))
sales_comparison.plot(kind="bar")

plt.title("Total Sales Comparison")
plt.ylabel("Revenue")
plt.xlabel("Brand")

plt.xticks(rotation=45)
plt.show()
```

```python
data = {
    "Brand": np.random.choice(["Levi's", "Calvin Klein"], 1000),
    "Revenue": np.random.randint(50, 500, 1000),
    "Cost": np.random.randint(20, 300, 1000),
    "Customer_Age": np.random.randint(18, 65, 1000),
    "Gender": np.random.choice(["Male", "Female"], 1000),
    "Income_Level": np.random.choice(["Low", "Medium", "High"], 1000),
    "Year": np.random.choice([2020,2021,2022,2023,2024], 1000)
}

df = pd.DataFrame(data)

df["Profit"] = df["Revenue"] - df["Cost"]
df["Profit_Margin"] = (df["Profit"] / df["Revenue"]) * 100

df.to_csv("fashion_sales.csv", index=False)

df.head()
```
