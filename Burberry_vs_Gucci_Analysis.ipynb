```python
# Step 2: Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display plots in notebook
%matplotlib inline

# style
sns.set(style="whitegrid")
```

```python
# Step 3: Create dataset

data = {
    "Company": ["Burberry", "Burberry", "Burberry",
                "Gucci", "Gucci", "Gucci"],

    "Year": [2023, 2024, 2025,
             2023, 2024, 2025],

    # revenue in billion USD (example realistic values)
    "Revenue_Billion_USD": [3.1, 2.9, 3.0,
                            10.5, 9.8, 9.2],

    # profit margin %
    "Profit_Margin_%": [16.0, 15.0, 14.5,
                        19.0, 18.5, 18.0],

    # demographics age distribution %
    "Age_18_25_%": [18, 20, 22,
                    25, 27, 29],

    "Age_26_35_%": [32, 34, 35,
                    38, 39, 40],

    "Age_36_50_%": [30, 28, 26,
                    24, 23, 22],

    "Age_50_plus_%": [20, 18, 17,
                      13, 11, 9],

    # gender distribution
    "Male_%": [45, 46, 47,
               42, 43, 44],

    "Female_%": [55, 54, 53,
                 58, 57, 56],

    # regional distribution
    "Europe_%": [35, 34, 36,
                 30, 29, 28],

    "Asia_%": [40, 41, 39,
               45, 46, 47],

    "America_%": [25, 25, 25,
                  25, 25, 25]
}

df = pd.DataFrame(data)

df
```

```python
# View first rows
df.head()
```

```python
# Dataset summary
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# revenue comparison plot

plt.figure(figsize=(10,6))

for company in df['Company'].unique():
    subset = df[df['Company'] == company]
    plt.plot(subset['Year'], subset['Revenue_Billion_USD'], marker='o', label=company)

plt.title("Revenue Comparison: Burberry vs Gucci")
plt.xlabel("Year")
plt.ylabel("Revenue (Billion USD)")
plt.legend()
plt.show()
```

```python
# profit margin comparison

plt.figure(figsize=(10,6))

sns.barplot(data=df, x="Year", y="Profit_Margin_%", hue="Company")

plt.title("Profit Margin Comparison")
plt.ylabel("Profit Margin %")
plt.show()
```

```python
# average demographics by company

demo_cols = ["Age_18_25_%", "Age_26_35_%", "Age_36_50_%", "Age_50_plus_%"]

demo_avg = df.groupby("Company")[demo_cols].mean()

demo_avg
```

```python
# plot demographics

demo_avg.T.plot(kind="bar", figsize=(10,6))

plt.title("Customer Age Demographics Comparison")
plt.ylabel("Percentage %")
plt.show()
```

```python
gender_cols = ["Male_%", "Female_%"]

gender_avg = df.groupby("Company")[gender_cols].mean()

gender_avg
```

```python
gender_avg.T.plot(kind="bar", figsize=(8,5))

plt.title("Gender Demographics")
plt.ylabel("%")
plt.show()
```

```python
region_cols = ["Europe_%", "Asia_%", "America_%"]

region_avg = df.groupby("Company")[region_cols].mean()

region_avg
```

```python
region_avg.T.plot(kind="bar", figsize=(8,5))

plt.title("Regional Demographics")
plt.ylabel("%")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Revenue_Billion_USD",
    y="Profit_Margin_%",
    hue="Company",
    s=150
)

plt.title("Revenue vs Profit Margin")
plt.show()
```

```python
# calculate averages

summary = df.groupby("Company").agg({
    "Revenue_Billion_USD": "mean",
    "Profit_Margin_%": "mean"
})

summary
```

```python
df.to_csv("burberry_gucci_dataset.csv", index=False)

print("Dataset exported successfully")
```
